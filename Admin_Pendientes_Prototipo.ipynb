# **Listas**

```python
pendientes_grandes = []
pendientes_medianos = []
pendientes_pequenos = []
pendientes_express = []
pendientes_periodicos = []
pendientes_diarios = []
```

# **Funcion de clasificacion de pendientes**

```python
print("Admin Pendientes v0.1")

def clasificar(pendiente):
  print("\nClasificando pendiente -> ",pendiente)
  tiempo = input("\nCuanto tiempo calculas que te tomara ese pendiente? \nColoca e indica el tiempo en minutos: ")
  tiempo = int(tiempo)
  prioridad = input("\nIngresa prioridad del pendiente \n -Urgente \n -Media \n -Baja \n Prioridad: ")
  prioridad.lower()


  periodico = input("\nTu pendiente es periodico? (Si o No): ")
  periodico = periodico.lower()

  if periodico == "si":
    diario = input("\nTu pendiente debes hacerlo diario? (Si o No): ")
    diario = diario.lower()

    if diario == "si":
      clas_en_listas(pendiente,tiempo,prioridad,False,diario,0)
    elif diario == "no":
      dias = input("\nCada cuento debes repetir el pendiente? (Indica en dias): ")
      dias = int(dias)
      clas_en_listas(pendiente,tiempo,prioridad,False,"no",dias)
  elif periodico == "no":
    clas_en_listas(pendiente,tiempo,prioridad,False)
```

```python
def clas_en_listas(pendiente,tiempo,prioridad,realizado,diario="no",dias=0):

    if diario == "si":
      pendientes_diarios.append((pendiente,tiempo,prioridad,realizado))
    elif dias != 0:
      pendientes_periodicos.append((pendiente,tiempo,prioridad,realizado,dias))
    else:

      if tiempo <= 15:
        pendientes_express.append((pendiente,tiempo,prioridad,realizado))
      elif tiempo <= 60:
        pendientes_pequenos.append((pendiente,tiempo,prioridad,realizado))
      elif tiempo <= 180:
        pendientes_medianos.append((pendiente,tiempo,prioridad,realizado))
      elif tiempo > 180:
        pendientes_grandes.append((pendiente,tiempo,prioridad,realizado))
```

# **Agregar nuevo pendiente**

```python
def agregar_pendiente():
  pendiente = input("\nCual es el pendiente que quieres agregar: ")
  clasificar(pendiente)
```

# **Ordenar y dar ideas de que pendientes se pueden resolver dependiendo del tiempo que se tiene**

```python
def lista_pendietes_posibles():
  tiempo_libre = input("\nCuanto tiempo tienes? (Indica en minutos): ")
  tiempo_libre = int(tiempo_libre)
  print("\n... Veamos que pendientes podrias hacer c: ...")

  pendientes_candidatosUrgente = []
  pendientes_candidatosMedia = []
  pendientes_candidatosBaja = []

  for posible_pendiente in pendientes_express:
    print("if ",posible_pendiente[1]," <= ",tiempo_libre," and ",posible_pendiente[3]," == False")
    print("if ",posible_pendiente[1]<=tiempo_libre," and ",posible_pendiente[3] == False)
    print(len(posible_pendiente[3]))

    if posible_pendiente[1] <= tiempo_libre and posible_pendiente[3] == "False":
      if posible_pendiente[2] == "urgente":
        pendientes_candidatosUrgente.append(posible_pendiente)
      elif posible_pendiente[2] == "media":
        pendientes_candidatosMedia.append(posible_pendiente)
      elif posible_pendiente[2] == "baja":
        pendientes_candidatosBaja.append(posible_pendiente)

  for posible_pendiente in pendientes_pequenos:
    if posible_pendiente[1] <= tiempo_libre and posible_pendiente[3] == False:

      if posible_pendiente[2] == "urgente":
        pendientes_candidatosUrgente.append(posible_pendiente)
      elif posible_pendiente[2] == "media":
        pendientes_candidatosMedia.append(posible_pendiente)
      elif posible_pendiente[2] == "baja":
        pendientes_candidatosBaja.append(posible_pendiente)

  for posible_pendiente in pendientes_medianos:
    if posible_pendiente[1] <= tiempo_libre and posible_pendiente[3] == False:

      if posible_pendiente[2] == "urgente":
        pendientes_candidatosUrgente.append(posible_pendiente)
      elif posible_pendiente[2] == "media":
        pendientes_candidatosMedia.append(posible_pendiente)
      elif posible_pendiente[2] == "baja":
        pendientes_candidatosBaja.append(posible_pendiente)

  for posible_pendiente in pendientes_grandes:
    if posible_pendiente[1] <= tiempo_libre and posible_pendiente[3] == False:

      if posible_pendiente[2] == "urgente":
        pendientes_candidatosUrgente.append(posible_pendiente)
      elif posible_pendiente[2] == "media":
        pendientes_candidatosMedia.append(posible_pendiente)
      elif posible_pendiente[2] == "baja":
        pendientes_candidatosBaja.append(posible_pendiente)

  for posible_pendiente in pendientes_diarios:
    if posible_pendiente[1] <= tiempo_libre and posible_pendiente[3] == False:

      if posible_pendiente[2] == "urgente":
        pendientes_candidatosUrgente.append(posible_pendiente)
      elif posible_pendiente[2] == "media":
        pendientes_candidatosMedia.append(posible_pendiente)
      elif posible_pendiente[2] == "baja":
        pendientes_candidatosBaja.append(posible_pendiente)
  print(pendientes_candidatosUrgente)
  print(pendientes_candidatosMedia)
  print(pendientes_candidatosBaja)
  print("\nPuedes hacer (ordenados por prioridad): \n")
  for p in pendientes_candidatosUrgente:
    print("- ",p[0]," Tiempo: ",p[1]," Prioridad: ",p[2])
  for p in pendientes_candidatosMedia:
    print("- ",p[0]," Tiempo: ",p[1]," Prioridad: ",p[2])
  for p in pendientes_candidatosBaja:
    print("- ",p[0]," Tiempo: ",p[1]," Prioridad: ",p[2])
```

# **Procesar lista de pendientes para clasificarlos**

```python
def procesar_pendientes_Txt():
  nombreArchivo = input("\nIngresa el nombre del archivo: ")
  with open(nombreArchivo, "r", encoding="utf-8") as archivo:
    print("Analizando archivo...\n")
    for renglon in archivo:
      pendiente = renglon.strip()
      clasificar(pendiente)

```

# **Guardar datos en archivo**

```python
def guardar_datos():
  with open("pendientesData.txt", "w") as archivo:
    for pendienteInfo in pendientes_express:
      archivo.write(f"{pendienteInfo[0]}#{pendienteInfo[1]}#{pendienteInfo[2]}#{pendienteInfo[3]}")
    for pendienteInfo in pendientes_pequenos:
      archivo.write(f"{pendienteInfo[0]}#{pendienteInfo[1]}#{pendienteInfo[2]}#{pendienteInfo[3]}")
    for pendienteInfo in pendientes_medianos:
      archivo.write(f"{pendienteInfo[0]}#{pendienteInfo[1]}#{pendienteInfo[2]}#{pendienteInfo[3]}")
    for pendienteInfo in pendientes_grandes:
      archivo.write(f"{pendienteInfo[0]}#{pendienteInfo[1]}#{pendienteInfo[2]}#{pendienteInfo[3]}")
    for pendienteInfo in pendientes_diarios:
      archivo.write(f"{pendienteInfo[0]}#{pendienteInfo[1]}#{pendienteInfo[2]}#{pendienteInfo[3]}")
    for pendienteInfo in pendientes_periodicos:
      archivo.write(f"{pendienteInfo[0]}#{pendienteInfo[1]}#{pendienteInfo[2]}#{pendienteInfo[3]}#{pendienteInfo[4]}")
    print("\nDatos guardados con exito !! c:")
```

# **Cargar pendientes**

```python
def cargar_pendientes():
  nombreArchivo = input("\nIngresa el nombre del archivo para cargar tus datos: ")
  with open(nombreArchivo, "r") as archivo:
    for renglon in archivo:
      renglon = renglon.strip("\n")
      pendienteData = renglon.split("#")
      if len(pendienteData) == 4:
        clas_en_listas(pendienteData[0],int(pendienteData[1]),pendienteData[2],pendienteData[3])
      elif len(pendienteData) == 5:
        clas_en_listas(pendienteData[0],int(pendienteData[1]),pendienteData[2],pendienteData[3],int(pendienteData[4]))
  print("\nCarga completada c:")
```
